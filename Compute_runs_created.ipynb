First cut at figuring runs created

```python
RCest = []

num_batters = len(batting_df)
for x in range(num_batters):
    # assign to variables to make the formulas easier to write and read
    H    = batting_df.iloc[x, 8]
    BB   = batting_df.iloc[x, 15]
    HBP  = batting_df.iloc[x, 18] 
    CS   = batting_df.iloc[x, 14]
    GIDP = batting_df.iloc[x, 21]
    IBB  = batting_df.iloc[x, 17]
    SB   = batting_df.iloc[x, 13]
    SH   = batting_df.iloc[x, 19]
    SF   = batting_df.iloc[x, 20]
    K    = batting_df.iloc[x, 16]
    DBL  = batting_df.iloc[x, 9]
    TPL  = batting_df.iloc[x, 10]
    HR   = batting_df.iloc[x, 11]
    AB   = batting_df.iloc[x, 6]
    
    A = H + BB + HBP - CS - GIDP
    B = (.24 * (BB - IBB + HBP)) + (.62 * SB) + (.5 * (SH + SF)) + (H + DBL + (2 * TPL) + (3 * HR))
    C = AB + BB + HBP + SH + SF
    
    A1 = C * 2.4 + A
    B1 = C * 3.0 + B
    
    if (C > 0):
        RC = ((A1 * B1)/(C * 9) - (C * .9))
    else: 
        RC = 0
    if (RC < 0):
        RC = 0
    RCest.append(RC)
    
batting_df["RCest"] = RCest
display(batting_df)
```
