```python
import pandas as pd
```

```python
batting_df = pd.read_csv("Batting_subset.csv")
```

```python
display(batting_df)
```

First cut at figuring runs created

```python
RCest = []

num_batters = len(batting_df)
for x in range(num_batters):
    # assign to variables to make the formulas easier to write and read
    H    = batting_df.iloc[x, 7]
    BB   = batting_df.iloc[x, 14]
    HBP  = batting_df.iloc[x, 17] 
    CS   = batting_df.iloc[x, 13]
    GIDP = batting_df.iloc[x, 20]
    IBB  = batting_df.iloc[x, 16]
    SB   = batting_df.iloc[x, 12]
    SH   = batting_df.iloc[x, 18]
    SF   = batting_df.iloc[x, 19]
    K    = batting_df.iloc[x, 15]
    DBL  = batting_df.iloc[x, 8]
    TPL  = batting_df.iloc[x, 9]
    HR   = batting_df.iloc[x, 10]
    AB   = batting_df.iloc[x, 5]
    
    A = H + BB + HBP - CS - GIDP
    B = (.24 * (BB - IBB + HBP)) + (.62 * SB) + (.5 * (SH + SF)) + (H + DBL + (2 * TPL) + (3 * HR))
    C = AB + BB + HBP + SH + SF
    
    A1 = C * 2.4 + A
    B1 = C * 3.0 + B
    
    if (C > 0):
        RC = ((A1 * B1)/(C * 9) - (C * .9))
    else: 
        RC = 0
    if (RC < 0):
        RC = 0
    RCest.append(RC)
    
batting_df["RCest"] = RCest
display(batting_df)
```

Next, we will adjust these estimates by the ratio of their sum to the actual team runs scored

```python
# sum up both runs and runs created estimate for each team/year
R_est_df = batting_df.groupby(["teamID", "yearID"], as_index=False).sum()
display (R_est_df)
```

```python
R_est_df["r_ratio"] = R_est_df["R"] / R_est_df["RCest"]
xyz=['teamID',"yearID","r_ratio"]
R_est2_df = R_est_df[xyz]

display(R_est2_df)
```

```python
batting_df = pd.merge(batting_df, R_est2_df, on=("teamID", "yearID"))
display(batting_df)
```

```python
batting_df["RC"] = batting_df["RCest"] * batting_df["r_ratio"]     
display(batting_df)
```

```python
batting_df.to_csv("batting_subset2.csv", index=False, header=True)
```
