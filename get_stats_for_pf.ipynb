```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
from splinter import Browser
```

```python
# ESPN-designated team labels - they seem to correspond with teams consistently
teams = ["team-10-1", "team-10-2", "team-10-3", "team-10-4", "team-10-5", 
         "team-10-6", "team-10-7", "team-10-8", "team-10-9", "team-10-10",
         "team-10-11", "team-10-12", "team-10-13", "team-10-14", "team-10-15", 
         "team-10-16", "team-10-17", "team-10-18", "team-10-19", "team-10-20",
         "team-10-21", "team-10-22", "team-10-23", "team-10-24", "team-10-25", 
         "team-10-26", "team-10-27", "team-10-28", "team-10-29", "team-10-30" ]
```

```python
# this obtains runs and home runs made @ home
x_yr = [];
x_team_id = [];
x_team = [];
x_runs33f = [];
x_hr33f = [];
x_runs34f = [];
x_hr34f = [];
x_runs33a = [];
x_hr33a = [];
x_runs34a = [];
x_hr34a = [];

for year in range(2002, 2019):
    strYear = str(year)
    
    # find (batting) runs and home runs at home
    url = "http://www.espn.com/mlb/stats/team/_/stat/batting/year/" + strYear + "/split/33"
    response = requests.get(url)
    soup = bs(response.text, 'html.parser')
    
    for team in teams:
        result = soup.find("tr", class_=team)
        stats_01   = result.find_all("td")
        x_yr.append(year)
        x_team_id.append(team)
        x_team.append(stats_01[1].text)
        x_runs33f.append(stats_01[4].text)
        x_hr33f.append(stats_01[8].text)
        
    # find (battiong) runs and home runs on the road
    url = "http://www.espn.com/mlb/stats/team/_/stat/batting/year/" + strYear + "/split/34"
    response = requests.get(url)
    soup = bs(response.text, 'html.parser')
    
    for team in teams:
        result = soup.find("tr", class_=team)
        stats_01   = result.find_all("td")
        x_runs34f.append(stats_01[4].text)
        x_hr34f.append(stats_01[8].text)

        
    # find runs against (pitching) at home
    url = "http://www.espn.com/mlb/stats/team/_/stat/pitching/year/" + strYear + "/split/33"
    response = requests.get(url)
    soup = bs(response.text, 'html.parser')
    
    for team in teams:
        result = soup.find("tr", class_=team)
        stats_01   = result.find_all("td")
        x_runs33a.append(stats_01[12].text)
        
    # find home runs against (pitching) at home
    url = "http://www.espn.com/mlb/stats/team/_/stat/pitching/year/" + strYear + "/split/33/type/expanded"
    response = requests.get(url)
    soup = bs(response.text, 'html.parser')
    
    for team in teams:
        result = soup.find("tr", class_=team)
        stats_01   = result.find_all("td")
        x_hr33a.append(stats_01[13].text)
        
    # find runs against (pitching) on the road
    url = "http://www.espn.com/mlb/stats/team/_/stat/pitching/year/" + strYear + "/split/34"
    response = requests.get(url)
    soup = bs(response.text, 'html.parser')
    
    for team in teams:
        result = soup.find("tr", class_=team)
        stats_01   = result.find_all("td")
        x_runs34a.append(stats_01[12].text)
        
    # find home runs against (pitching) on road
    url = "http://www.espn.com/mlb/stats/team/_/stat/pitching/year/" + strYear + "/split/34/type/expanded"
    response = requests.get(url)
    soup = bs(response.text, 'html.parser')
    
    for team in teams:
        result = soup.find("tr", class_=team)
        stats_01   = result.find_all("td")
        x_hr34a.append(stats_01[13].text)
        
        
```

```python
print (len(x_yr), len(x_team_id), len(x_team))
```

```python
# weave data from above into a data frame
df_pf_stats = pd.DataFrame({"year": x_yr, "team_id" : x_team_id, "team": x_team, 
                            "runs at home for": x_runs33f, "hr at home for": x_hr33f,
                            "runs on road for": x_runs34f, "hr on road for": x_hr34f,
                            "runs at home against": x_runs33a, "hr at home against": x_hr33a,
                            "runs on road against": x_runs34a, "hr on road against": x_hr34a })
df_pf_stats.head(20)
```

```python
df_pf_stats.to_csv("park_factor_stats.csv", index=False, header=True)
```
