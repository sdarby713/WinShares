```python
import pandas as pd
import math
```

```python
park_factor_df = pd.read_csv("park_factor_stats.csv")
display(park_factor_df)
```

```python
# since we built this dataframe file, we know that all data is present.  any field could be counted
df_length = park_factor_df.year.count() - 1

# N = Number of teams in League
# With the number of interleague games being played, we will regard "league" as the Major Leagues
# we know this to be constant through the period in scope
N = 30;

```

```python
# total runs and homeruns for league will be needed
park_factor_groupby_year = park_factor_df.groupby('year')
league_runs = park_factor_groupby_year["runs at home for"].sum() + park_factor_groupby_year["runs at home against"].sum()
league_runs_df = pd.DataFrame(league_runs)

league_hrs  = park_factor_groupby_year["hr at home for"].sum() + park_factor_groupby_year["hr at home against"].sum()
league_hrs_df = pd.DataFrame(league_hrs)

display(league_runs_df)
runs_league = league_runs.loc[2006]
runs_league
```

Compute the park factors for all teams / all years

```python
year_list = []
team_list = []
PRFa_list = []
PHRFa_list = []
PNHRFa_list = []

for i in range(2,df_length):
# for i in range(509,510):    # for testing
    year = park_factor_df.iloc[i, 0]
    team = park_factor_df.iloc[i, 2]
    curr_park = park_factor_df.iloc[i, 11]
    
    tot_weight = 4
    tot_r_home  = (park_factor_df.iloc[i, 3] + park_factor_df.iloc[i, 7])*4
    tot_hr_home = (park_factor_df.iloc[i, 4] + park_factor_df.iloc[i, 8])*4
    tot_r_road  = (park_factor_df.iloc[i, 5] + park_factor_df.iloc[i, 9])*4
    tot_hr_road = (park_factor_df.iloc[i, 6] + park_factor_df.iloc[i, 10])*4
        
    j = i - 2
    if (park_factor_df.iloc[j, 11] == curr_park):
        tot_weight  += 1
        tot_r_home  += (park_factor_df.iloc[j, 3] + park_factor_df.iloc[j, 7])
        tot_hr_home += (park_factor_df.iloc[j, 4] + park_factor_df.iloc[j, 8])
        tot_r_road  += (park_factor_df.iloc[j, 5] + park_factor_df.iloc[j, 9])
        tot_hr_road += (park_factor_df.iloc[j, 6] + park_factor_df.iloc[j, 10])

    j = i - 1
    if (park_factor_df.iloc[j, 11] == curr_park):
        tot_weight  += 1
        tot_r_home  += (park_factor_df.iloc[j, 3] + park_factor_df.iloc[j, 7])
        tot_hr_home += (park_factor_df.iloc[j, 4] + park_factor_df.iloc[j, 8])
        tot_r_road  += (park_factor_df.iloc[j, 5] + park_factor_df.iloc[j, 9])
        tot_hr_road += (park_factor_df.iloc[j, 6] + park_factor_df.iloc[j, 10])
      
    j = i + 1
    if (j <= df_length):    # don't go past end of data
        if (park_factor_df.iloc[j, 11] == curr_park):
            tot_weight  += 1
            tot_r_home  += (park_factor_df.iloc[j, 3] + park_factor_df.iloc[j, 7])
            tot_hr_home += (park_factor_df.iloc[j, 4] + park_factor_df.iloc[j, 8])
            tot_r_road  += (park_factor_df.iloc[j, 5] + park_factor_df.iloc[j, 9])
            tot_hr_road += (park_factor_df.iloc[j, 6] + park_factor_df.iloc[j, 10])
              
    j = i + 2
    if (j <= df_length):    # don't go past end of data
        if (park_factor_df.iloc[j, 11] == curr_park):
            tot_weight  += 1
            tot_r_home  += (park_factor_df.iloc[j, 3] + park_factor_df.iloc[j, 7])
            tot_hr_home += (park_factor_df.iloc[j, 4] + park_factor_df.iloc[j, 8])
            tot_r_road  += (park_factor_df.iloc[j, 5] + park_factor_df.iloc[j, 9])
            tot_hr_road += (park_factor_df.iloc[j, 6] + park_factor_df.iloc[j, 10])
            
    PRF  = (tot_r_home  / tot_weight) / (tot_r_road  / tot_weight)
    PRFa = (( N - 1) * PRF + ( N - PRF)) / (2 * (N - 1))
    PHRF = (tot_hr_home / tot_weight) / (tot_hr_road / tot_weight)
    PHRFa = (( N - 1) * PHRF + ( N - PHRF)) / (2 * (N - 1))
    

    runs_league = league_runs.loc[year]
    hr_league   = league_hrs.loc[year]
    factor_a    = 1.5 * (hr_league/runs_league)
    factor_b    = 1 - factor_a
    PNHRFa      = math.sqrt((PRFa - (factor_a * PHRFa)) / factor_b)
    
    year_list.append(year)
    team_list.append(team)
    PRFa_list.append(PRFa)
    PHRFa_list.append(PHRFa)
    PNHRFa_list.append(PNHRFa)
    
# with that done, assemble our park factor df
dict1 = {"year" : year_list,
         "team" : team_list,
         "PRFa" : PRFa_list,
         "PHRFa" : PHRFa_list,
         "PNHRFa" : PNHRFa_list}
pf_data_df = pd.DataFrame(dict1)
        
    
    
```
